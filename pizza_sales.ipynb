```python
select sum(total_price) / count(distinct (order_id)) as 'avg_order_value' from 'pizza_sales.csv'
```

```python
select count(distinct(order_id)) as 'total_orders' from 'pizza_sales.csv'
```

```python
select sum(quantity) / count(distinct(order_id)) as 'avg_pizza_per_order' from 'pizza_sales.csv'
```

```python
SELECT dayname(order_date) AS order_day, COUNT(DISTINCT order_id) AS total_orders 
FROM 'pizza_sales.csv'
GROUP BY dayname(order_date);
```

```python
SELECT dayname(order_date) AS order_day, 
sum( quantity) AS total_quantity
FROM 'pizza_sales.csv'
GROUP BY dayname(order_date);
```

```python
select monthname(order_date) as month_name, 
count(distinct order_id) as total_orders
from 'pizza_sales.csv'
group by monthname(order_date) 
order by total_orders desc;
```

```python
select pizza_category, 
sum(total_price) as 'total_sales',
(sum(total_price) * 100 / (select sum(total_price) from 'pizza_sales.csv')) as '%_sales_by_pizza_category'
from 'pizza_sales.csv'
group by pizza_category;
```

```python
select pizza_size, 
sum(total_price) as 'total_sales',
(sum(total_price) * 100 / (select sum(total_price) from 'pizza_sales.csv')) as '%_sales_by_pizza_size'
from 'pizza_sales.csv'
group by pizza_size;
```
